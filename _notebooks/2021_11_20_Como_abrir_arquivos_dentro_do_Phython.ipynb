Definimos uma variável para ser acessada com mais facilidade, chamada de "caminho_base", contendo o diretório dos aquivos na raiz da onde ficam os upload de dentro do Colab.

````python
caminho_base = './{name_file}'
````

Para realizar leturas de arquivos do tipo csv, o Pandas necessita declarar o nome do arquivo como variável, seguido da execussão do comando "read_csv(arquivo a ser lido.format(name_file=name_file), sep=',', decimal='.' )".

Isso deverá trazer o arquivo lido em formato de tabela. 

````python
## Leitura csv 
# Doc read_csv - https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
name_file = 'csv_file.csv'
pd.read_csv(caminho_base.format(name_file=name_file), sep=',', decimal='.' )
````

Para Leitura XLSX, o Pandas necessita declarar o nome do arquivo como variável, seguido da execussão do comando "pd.read_excel(arquivo a ser lido.format(name_file=name_file))".

Isso deverá trazer o arquivo lido em formato de tabela.

````python
## Leitura XLSX
# Doc read_excel https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html
name_file = 'xlsx_file.xlsx'
pd.read_excel(caminho_base.format(name_file=name_file))
````

Para Leitura txt, o Pandas necessita declarar o nome do arquivo como variável, seguido ou da execussão do comando "f = open(arquivo a ser lido.format(name_file=name_file), "r")
print(f.read())
f.close()". 

Ou para um método com "with", podemos usar o comando "print('\n \033[1m Titulo a inserir \n \033[0m')
with open(arquivo a ser lido.format(name_file=name_file), 'r') as f:
lines = f.read()
print(lines)"

Isso deverá trazer o arquivo lido em formato de texto.

````python
## Leitura txt
name_file='text_file.txt'
##Leitura com with

## Leitura sem with
f = open(caminho_base.format(name_file=name_file), "r")
print(f.read())
f.close()

## Leitura com with
print('\n \033[1m Leitura com With \n \033[0m')
with open(caminho_base.format(name_file=name_file), 'r') as f:
    lines = f.read()
    print(lines)
````

Para abrir Imagens, devemos importar uma bibliota chamada "PIL", o comando "Image", seguida do nome do arquivo jpg como variável e do comando "im = Image.open(caminho_base.format(name_file=name_file), 'r')". Além do comando "display(im) im.close()" para tornar visualizavel em qualquer visualizador. 


````python
# Leitura Imagem

from PIL import Image
  
name_file='image_file.jpg'
# open method used to open different extension image file
im = Image.open(caminho_base.format(name_file=name_file), 'r') 
  
# This method will show image in any image viewer 
display(im)
im.close()
````
